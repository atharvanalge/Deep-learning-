```python
import numpy as np
import matplotlib.pyplot as plt

# Define the gradient descent algorithm
def gradient_descent(W, b, alpha, dW, db):
  """
  Gradient Descent Algorithm:
  W : Weight term
  b : bias term
  alpha : learning rate
  dW : partial derivative of W value
  db : partial derivative of b value
  """

  # Update the weight and bias terms using the gradient descent algorithm
  W = W - alpha * dW
  b = b - alpha * db

  return W, b

# Define the cost function
def cost(W, b, y, m):
  """
  Calculate the cost function for the given weight and bias terms.

  Args:
    W: A numpy array of shape (n_features, 1), where n_features is the number of features.
    b: A numpy array of shape (1, 1).
    y: A numpy array of shape (n_samples, 1), where n_samples is the number of training examples.
    m: The number of training examples.

  Returns:
    A numpy array of shape (1, 1), where the first element is the cost function value.
  """

  # Calculate the predicted values
  y_pred = np.dot(W.T, X.T) + b

  # Calculate the cost function value
  cost_value = 1 / m * np.sum((y_pred - y.T) ** 2)

  return cost_value

# Load the training data
X = np.array([i for i in range(5)]).reshape(-1, 1)
y = (2 * np.array([i for i in range(5)]) + np.random.random(5) * 1).reshape(-1, 1)

# Initialize the weight and bias terms
W = np.array([0]).reshape(-1, 1)
b = np.array([0]).reshape(-1, 1)

# Set the number of iterations and the learning rate
iterations = 100
alpha = 0.01

# Initialize the cost history
COST = []

# Perform gradient descent
for i in range(iterations):

  # Calculate the partial derivatives of the cost function with respect to the weight and bias terms
  dW = 1 * np.dot(np.dot(W.T, X.T) + b - y.T, X)
  db = 1 * np.sum(np.dot(W.T, X.T) + b - y.T)

  # Update the weight and bias terms
  W, b = gradient_descent(W, b, alpha, dW, db)

  # Calculate the cost function value
  cost_value = cost(W, b, y, 5)

  # Append the cost function value to the cost history
  COST.append(cost_value)

  # Print the current weight and bias terms
  print(W, b)
```

```python
import numpy as np

def gradient_descent(W, b, alpha, dW, db):
    W = W - alpha * dW
    b = b - alpha * db
    return W, b

def cost(W, b, X, y, m):
    return 1 / (2 * m) * np.sum((np.dot(X, W) + b - y) ** 2)

X = np.array([i for i in range(5)]).reshape(-1, 1)
arr1 = np.random.random(5)
y = (2 * np.array([i for i in range(5)]) + arr1 * 1).reshape(-1, 1)

W = np.array([0]).reshape(-1, 1)
b = np.array([0]).reshape(-1, 1)

iterations = 100
alpha = 0.01

COST = []
batch_size = 2  # Set the batch size, e.g., 2 for this example

m = X.shape[0]  # Number of data points

for i in range(iterations):
    for j in range(0, m, batch_size):
        X_batch = X[j:j + batch_size]
        y_batch = y[j:j + batch_size]
        dW = 1/batch_size * np.dot(X_batch.T, (np.dot(X_batch, W) + b - y_batch))
        db = 1/batch_size * np.sum(np.dot(X_batch, W) + b - y_batch)
        W, b = gradient_descent(W, b, alpha, dW, db)
    COST.append(cost(W, b, X, y, batch_size))
    print("Iteration", i + 1, " - W:", W, " b:", b)

print("Final W and b:", W, b)
```

```python
import numpy as np

def momentum(X, y, batch_size, W, b, alpha, beta, dW, db, iterations):
    VdW = 0
    Vdb = 0

    for i in range(iterations):
        for j in range(0, X.shape[0], batch_size):
            X_batch = X[j:j+batch_size]
            y_batch = y[j:j+batch_size]

            dW = 1/batch_size * np.dot(np.dot(W.T, X_batch.T) + b - y_batch.T, X_batch)
            db = 1/batch_size * np.sum(np.dot(W.T, X_batch.T) + b - y_batch.T)

            VdW = beta * VdW + (1 - beta) * dW
            Vdb = beta * Vdb + (1 - beta) * db

            W = W - alpha * VdW
            b = b - alpha * Vdb

    return W, b

def cost(W, b, X, y, m):
    return 1 / (2 * m) * np.sum((np.dot(X, W) + b - y) ** 2)

X = np.array([i for i in range(5)]).reshape(-1, 1)
arr1 = np.random.random(5)
y = (2 * np.array([i for i in range(5)]) + arr1 * 1).reshape(-1, 1)

W = np.array([0]).reshape(-1, 1)
b = np.array([0]).reshape(-1, 1)

COST = []
batch_size = 2
iterations = 100
alpha = 0.09
beta = 0.1

for i in range(iterations):
    W, b = momentum(X, y, batch_size, W, b, alpha, beta, dW, db, iterations)
    COST.append(cost(W, b, X, y, batch_size))
    print("Iteration", i + 1, " - W:", W, " b:", b)

print("Final W and b:", W, b)
```
