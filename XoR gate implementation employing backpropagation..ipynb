# ATHARVA NALGE

```python
import numpy as np

# Define the sigmoid function and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Input data (XOR truth table)
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])

# Output data
Y = np.array([[0], [1], [1], [0]])

# Initialize weights and biases
input_size = 2
hidden_size = 4
output_size = 1

# Weights and biases for the input and hidden layer
w_input_hidden = np.random.uniform(size=(input_size, hidden_size))
b_input_hidden = np.zeros((1, hidden_size))

# Weights and biases for the hidden and output layer
w_hidden_output = np.random.uniform(size=(hidden_size, output_size))
b_hidden_output = np.zeros((1, output_size))

learning_rate = 0.9
epochs = 1000

# Training the neural network
for epoch in range(epochs):
    # Forward propagation
    hidden_input = np.dot(X, w_input_hidden) + b_input_hidden
    hidden_output = sigmoid(hidden_input)
    output_layer_input = np.dot(hidden_output, w_hidden_output) + b_hidden_output
    output_layer_output = sigmoid(output_layer_input)

    # Calculate the loss
    loss = Y - output_layer_output

    # Backpropagation
    d_output = loss * sigmoid_derivative(output_layer_output)
    error_hidden_layer = d_output.dot(w_hidden_output.T)
    d_hidden_layer = error_hidden_layer * sigmoid_derivative(hidden_output)

    # Update weights and biases
    w_hidden_output += hidden_output.T.dot(d_output) * learning_rate
    w_input_hidden += X.T.dot(d_hidden_layer) * learning_rate
    b_hidden_output += np.sum(d_output, axis=0, keepdims=True) * learning_rate
    b_input_hidden += np.sum(d_hidden_layer, axis=0, keepdims=True) * learning_rate

    # Print the predictions for this epoch
    print(f"Epoch {epoch+1} - Predicted Output:")
    print(output_layer_output)

# Testing the final trained neural network
hidden_input = np.dot(X, w_input_hidden) + b_input_hidden
hidden_output = sigmoid(hidden_input)
output_layer_input = np.dot(hidden_output, w_hidden_output) + b_hidden_output
output_layer_output = sigmoid(output_layer_input)

print("Final Predicted Output:")
print(output_layer_output)
```
