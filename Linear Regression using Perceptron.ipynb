```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
np.set_printoptions(suppress=True)
```

```python
x,y,coef=datasets.make_regression(n_samples=200, #number of samples
                                 n_features=1, #number of features m
                                 n_informative=1, #number of useful features y
                                 noise=10,# Bias and standard deviation of gussain noise
                                 coef=True, #True coefficient used to generate the data
                                 random_state=0) #Set for same data points for each turn
```

```python
fig=plt.figure(figsize=[30,10])#Create figure
plt.plot(x,y,'x')
plt.title('Training Data')
plt.show()
```

```python
def square(y,net):
    return (y-net)**2
```

```python
def train_perceptron(inputs, target, weights, learning_rate):
    net = inputs*weights
    error = square(y,net)
    new_weights = weights + learning_rate * error * inputs
    new_weights = new_weights.astype(float) 
    return new_weights, error
```

```python
weight=0.0
learning_rate=0.01
iterations =10
bias=0.0
mse_list=[]

for i in range(iterations):
    mse=0.0 #intialize the mean square
    for xi,yi in zip(x,y):
        y_out=weight*xi+bias #y_out actual#yi target
        mse += (y_out-yi)**2 #mse
        weight -= learning_rate*(y_out-yi)*xi # so here we did actual-target so in this equation we subtracted
        bias -=learning_rate*(y_out-yi)
    mse /=len(x) #Mean Squared error
    mse_list.append(mse)
    print("Iteration:{},MSE:{}".format(i+1,mse))
```

```python
fig=plt.figure(figsize=[30,10])#Create figure
plt.plot(mse_list)
plt.title('Train')
plt.show()
```
